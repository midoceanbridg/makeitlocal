```python
import requests
from pathlib import Path
import os
import pickle
```

```python
#loading block
mykey = open('spoonac/apikey.txt').read().strip()
ingiddata = pickle.load(open("bigoven/classifying_ing_aisle_meta_measure_id.pkl", 'rb'))
```

```python
ingiddata
names = ingiddata['ingredient'].tolist()
```

```python
outname = []
substitute = []
subsetnames = names[0:]
n = 0

for ing in subsetnames:
    params = {'apiKey': mykey, 'ingredientName': ing}
    response = requests.get('https://api.spoonacular.com/food/ingredients/substitutes', params=params)
    res = response.json()
    n += 1
    if res['status'] != 'failure':
        outname.append(res['ingredient'])
        substitute.append(res['substitutes'])
    else:
        outname.append(None)
        substitute.append(None)
    
```

```python
newdat = {'output name': outname, 'subsitute': substitute}
```

```python
ingiddata.insert(5, 'output name', outname)
ingiddata.insert(6, 'subsitute', substitute)
```
