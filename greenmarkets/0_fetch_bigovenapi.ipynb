### Grab data from BigOven API

```python
import requests
from time import sleep
import os
```

```python
mykey = open('bigoven/apikey.txt').read().strip()
```

```python
#get rrecipes by ID

mealstage = 'maindish'

def getlistrec(pg, rpp):
    params = {'include_primarycat': mealstage, 'pg': str(pg), 'rpp': str(rpp), 'api_key': mykey}   
    response = requests.get('https://api2.bigoven.com/recipes', params=params)
    return response.json()

params = {'api_key': mykey} 
    

def getrecipes():
    trackerfile = f'bigoven/{mealstage}/pagegotto.txt'
    pg, n = open(trackerfile).read().strip().split(',')
    pg, n = int(pg), int(n)
    rpp = 250
    while True:
        print(f'Getting list of recipes: {pg} , {rpp}')
        response = getlistrec(pg, rpp)
        
        for i, recipemeta in enumerate(response['Results']):
            if i < n: continue 
                
            id = recipemeta['RecipeID']
            response = requests.get(f'https://api2.bigoven.com/recipe/{id}', params=params)
            if "Message" in response.json(): 
                with open(f'bigoven/{mealstage}/regrab.txt', 'a') as fh:
                    fh.write(f'{pg},{i},{id}\n')
                continue
                
            with open(f'bigoven/{mealstage}/{mealstage}-{pg}-{i}-{id}.json', 'a') as fh:
                fh.write(response.text)
            with open(f'bigoven/{mealstage}/pagegotto.new', 'w') as fh:
                fh.write(f'{pg},{i}\n')
            os.rename(f'bigoven/{mealstage}/pagegotto.new', trackerfile)
        pg += 1
        n = 0

while True:
    try:
        getrecipes()
    except Exception as e:
        print(f'Error getting recipes: {e}')
        sleep(60*30)
    
    
    
    
    
```

```python
params = {'include_primarycat': 'maindish', 'pg': str(1), 'rpp': str(250), 'api_key': mykey}    #'number': str(number), 'type': 'main course', 'apiKey': mykey, 'offset': str(i*number), 'cuisine': cuisine}
response = requests.get('https://api2.bigoven.com/recipes', params=params)
```
