### Grab data from BigOven API

```python
import requests
from time import sleep
import os
```

```python

mykey = open('eatlocal/generated_data/BOapikey.txt').read().strip()
```

```python
#get recipes by ID

mealstage = 'maindish'

def getlistrec(pg, rpp):
    '''A function for getting a list of available recipe IDs'''
    params = {'include_primarycat': mealstage, 'pg': str(pg), 'rpp': str(rpp), 'api_key': mykey}   
    response = requests.get('https://api2.bigoven.com/recipes', params=params)
    return response.json()

params = {'api_key': mykey} 
    

def getrecipes():
    '''Grabs the list of recipe IDs and iteratively downloads each recipe'''
    trackerfile = f'bigoven/{mealstage}testing/pagegotto.txt' #opening a file to track the number of previously saved recipes
    pg, n = open(trackerfile).read().strip().split(',') 
    pg, n = int(pg), int(n)
    rpp = 250
    while True:
        print(f'Getting list of recipes: {pg} , {rpp}')
        response = getlistrec(pg, rpp) #getting the IDs 
        
        #go through list of IDs, request full recipe, save
        for i, recipemeta in enumerate(response['Results']):
            if i < n: continue 
                
            id = recipemeta['RecipeID']
            response = requests.get(f'https://api2.bigoven.com/recipe/{id}', params=params)
            if "Message" in response.json(): 
                with open(f'bigoven/{mealstage}testing/regrab.txt', 'a') as fh:
                    fh.write(f'{pg},{i},{id}\n')
                continue
                
                #saving block
            with open(f'bigoven/{mealstage}testing/{mealstage}-{pg}-{i}-{id}.json', 'a') as fh:
                fh.write(response.text)
            with open(f'bigoven/{mealstage}testing/pagegotto.new', 'w') as fh:
                fh.write(f'{pg},{i}\n')
            os.rename(f'bigoven/{mealstage}testing/pagegotto.new', trackerfile)
        pg += 1
        n = 0

while True: #Grab as many recipes as possible
    try:
        getrecipes()
    except Exception as e:
        print(f'Error getting recipes: {e}')
        sleep(60*30) #bigoven throttles by the hour, so when we hit our limit wait 30 mins and try again
    
    
    
    
    
```
